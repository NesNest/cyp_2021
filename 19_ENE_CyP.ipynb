# slicing (continuacion)
[inicio:alto:incremento]

notas.
1. el alto no es incluido.
2. el incremento se puede modificar.
3. soporta indexado alternativo (negativos).

```
frase = "como hacer a la gente mas positiva?, muy facil, has que pierdan un electron!" 
print(frase)
#imprimir "ace"
print(frase[6:9:1])
#valores por defecto 
print(frase[:5:2])
print(len(frase))
#incremento 
print(frase[::2]) # toda la frase
print(frase[-1::-1]) # inverso

nombre = "Juan Jose"
print(nombre[1::-1])

primer = nombre[0:4:]
print(primer)

print(nombre[-9::1])
```

# Metodos de la clase String (STR)

En python todo es un objeto (P.O.O)
eso significa que cada tipo de datos trae consigo metodos( funciones incluidas)

para identificar los metodos disponibles se puede usar una de las dos funciones 
1. dir(str) resumen de los metodos de esa clase 
2. help(str) informacion completa de los metodos de esa clase

```
nombre = "Néstor"
print(nombre)
print(dir(str))
```

```
print(nombre)
print(nombre.isupper())
nombre = "PEDRO FERNANDEZ"
print(nombre.capitalize())
print(nombre.endswith("DEZ"))
print(nombre.replace("E", "*"))
ind_e = nombre.find("E")
print(ind_e)
print(nombre)
print(nombre.replace("E","@", 3))
print(nombre.lower())
nombre = "jose"
print(nombre.upper())
nombre = "josé josé, el principe de la cancion"
print(nombre.count("josé"))
```

# metodos para procesar strings



```

# metodos para procesar strings
numeros  = "   2 ,   1 ,     3 ,  4 ,  5 ,  45 , 7     "
print(numeros)

# como los estraigo 
# 1. eliminar los espacios de al principio  y al final strip()
numeros = numeros.strip()
print(numeros)

help(str.strip)
help(str.split)

numeros = numeros.split(",")
print(numeros)
  

for indice in range ( len(numeros )):
  numeros[indice] = numeros[indice].strip()
  print(numeros[indice])
print(numeros)


# convetir a enteros 

for indice in range ( len(numeros )): 
  numeros[indice] = int(numeros[indice])
print(numeros)
  
```
